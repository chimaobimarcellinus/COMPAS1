```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 15]

# Convert quantities to percentages
total = sum(values)
percentages = [value / total * 100 for value in values]

# Plotting the bar chart with percentages
bars = plt.bar(categories, percentages, color='skyblue')

# Adding title and labels
plt.title("Bar Graph with Percentages")
plt.xlabel("Categories")
plt.ylabel("Percentage (%)")

# Format y-axis labels to show percentages
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

# Annotate each bar with its percentage value
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}%', 
             ha='center', va='bottom')

# Showing the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    'race': ['White', 'Black', 'Asian', 'Hispanic', 'White', 'Black', 'Asian'],
    'sex': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female']
}
df_unique_name = pd.DataFrame(data)

# Create the crosstab
crosstab = pd.crosstab(df_unique_name['race'], df_unique_name['sex'])

# Select the desired rows and column
subset = crosstab.iloc[0:5, [1]]

# Calculate percentages
subset_percentage = subset.div(subset.sum(axis=1), axis=0) * 100

# Plotting the bar chart with percentages
ax = subset_percentage.plot(kind="bar", color='skyblue', figsize=(10, 6))

# Adding title and labels
plt.title("Race and Sex Distribution", fontsize=15, fontweight='bold')
plt.xlabel("Race", fontsize=12)
plt.ylabel("Percentage (%)", fontsize=12)

# Annotate each bar with its percentage value formatted to 2 decimal places
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = {
    'race': ['A', 'B', 'C', 'D', 'E', 'A', 'B', 'C', 'D', 'E'],
    'sex': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'count': [10, 20, 30, 25, 15, 5, 10, 15, 20, 10]
}
df_unique_name = pd.DataFrame(data)

# Create a crosstab
plotting = pd.crosstab(df_unique_name.race, df_unique_name.sex)

# Filter to get the top 5 races
plotting = plotting.iloc[0:5]

# Choose a colormap
cmap = plt.get_cmap('tab20c')
categories = df_unique_name.race.unique()
colors = cmap(np.linspace(0, 1, len(categories)))

# Create a bar plot
ax = plotting.plot(kind='bar', color=colors)

# Calculate percentages for annotation
total_values = plotting.sum().sum()
percentages = (plotting / total_values * 100).values

# Annotate each bar with its percentage value
for container, percentage_column in zip(ax.containers, percentages.T):
    ax.bar_label(container, labels=[f'{x:.2f}%' for x in percentage_column], label_type='edge')

# Set labels and title
ax.set_ylabel('Count')
ax.set_title('Bar Graph with Percentage Annotations')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = {
    'race': ['A', 'B', 'C', 'D', 'E', 'A', 'B', 'C', 'D', 'E'],
    'sex': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'count': [10, 20, 30, 25, 15, 5, 10, 15, 20, 10]
}
df_unique_name = pd.DataFrame(data)

# Create a crosstab
plotting = pd.crosstab(df_unique_name.race, df_unique_name.sex).iloc[0:5]

# Choose a colormap
cmap = plt.get_cmap('tab20c')
categories = df_unique_name.race.unique()
colors = cmap(np.linspace(0, 1, len(categories)))

# External list of percentages (must match the order of the bars)
percentages = [
    [10.0, 5.0],  # A
    [20.0, 10.0],  # B
    [30.0, 15.0],  # C
    [25.0, 20.0],  # D
    [15.0, 10.0]   # E
]

# Create a bar plot
ax = plotting.plot(kind='bar', color=colors)

# Annotate each bar with its percentage value
for container, percentage_column in zip(ax.containers, percentages):
    ax.bar_label(container, labels=[f'{x:.2f}%' for x in percentage_column], label_type='edge')

# Set labels and title
ax.set_ylabel('Count')
ax.set_title('Bar Graph with Percentage Annotations')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data creation
np.random.seed(42)
time_index = pd.date_range(start='2023-01-01', periods=500, freq='T')  # Minute frequency
categories = np.random.choice(['A', 'B', 'C'], 500)
values = np.random.rand(500)

# Create DataFrame
df = pd.DataFrame({'Time': time_index, 'Category': categories, 'Value': values})

# Create crosstab
crosstab_data = pd.crosstab(df['Time'], df['Category'], values=df['Value'], aggfunc='sum').fillna(0)

# Plotting the crosstab data
plt.figure(figsize=(14, 7))
for category in crosstab_data.columns:
    plt.plot(crosstab_data.index, crosstab_data[category], label=category)

plt.title('Time Series Data from Crosstab')
plt.xlabel('Time')
plt.ylabel('Aggregated Value')
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for multiple DataFrames
data1 = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [10, 20, 30, 40]
}

data2 = {
    'Category': ['E', 'F', 'G', 'H'],
    'Value2': [15, 25, 35, 45]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Setting the index to 'Category' for easier plotting
df1.set_index('Category', inplace=True)
df2.set_index('Category', inplace=True)

# Creating subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plotting the first DataFrame
df1.plot(kind='bar', ax=axes[0], title='DataFrame 1')

# Plotting the second DataFrame
df2.plot(kind='bar', ax=axes[1], title='DataFrame 2')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'col1': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C', 'C'],
        'col2': ['X', 'Y', 'X', 'Y', 'X', 'X', 'X', 'Y', 'X', 'Y']}
df = pd.DataFrame(data)

# Create crosstab and normalize to get percentages
crosstab = pd.crosstab(df['col1'], df['col2'], normalize='index') * 100

# Plot the bar chart
ax = crosstab.plot(kind='bar', stacked=False)

# Annotate the bars with percentage values
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# Set y-axis to display percentages
ax.set_ylabel('Percentage')

# Display the plot
plt.show()
```
